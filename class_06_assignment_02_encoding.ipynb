#without encoding using loops

```python
from pandas.core.dtypes.common import is_numeric_dtype
```

```python



for column in df4.columns:
    something = df4[column].unique()
    if is_numeric_dtype(something):
        continue

    len = df4[column].nunique()
    i=0
    list_nominal = [] #list of nominal values in each column
    list_numeric = [] #list of numeric values
    for i in range(len):
        list_nominal.append(something[i])
        list_numeric.append(i)
    
    
    df4[column] = df[column].replace(list_nominal,list_numeric)
     
```
